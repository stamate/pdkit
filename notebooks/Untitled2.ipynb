```python
import os
os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"]="1"

import pandas as pd
import sys
sys.path.append('../pdkit')

from keras.layers import Dense, BatchNormalization, Dropout, Input
from keras.models import Model
from sklearn.preprocessing import StandardScaler
```

```python
train = pd.read_csv('/src/data/pdkit/pd_sound/speech_multiple/train_data.txt', header=None)

tr_x = train.iloc[:, 1:-2].values
tr_y = train.iloc[:, -1].values

test = pd.read_csv('/src/data/pdkit/pd_sound/speech_multiple/test_data.txt', header=None)

ts_x = test.iloc[:, 1:-1].values
ts_y = test.iloc[:, -1].values
```

```python
sc = StandardScaler()
sc.fit(tr_x)

tr_x = sc.transform(tr_x)
ts_x = sc.transform(ts_x)
```

```python
inputs = Input(shape=(26,))

dense = Dense(500, activation='relu')(inputs)
# dense = BatchNormalization()(dense)
dense = Dropout(0.3)(dense)

dense = Dense(500, activation='relu')(inputs)
# dense = BatchNormalization()(dense)
dense = Dropout(0.3)(dense)

dense = Dense(500, activation='relu')(inputs)
# dense = BatchNormalization()(dense)
dense = Dropout(0.3)(dense)

# dense = Dense(500, activation='relu')(dense)
# dense = BatchNormalization()(dense)
# # dense = Dropout(0.3)(dense)

# dense = Dense(500, activation='relu')(dense)
# # dense = BatchNormalization()(dense)
# dense = Dropout(0.3)(dense)

# dense = Dense(500, activation='relu')(dense)
# dense = BatchNormalization()(dense)
# # dense = Dropout(0.3)(dense)

output = Dense(2, activation='softmax')(dense)

model = Model(inputs=inputs, outputs=output)
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(tr_x, tr_y, batch_size=1, epochs=200, validation_data=[ts_x, ts_y])
```
